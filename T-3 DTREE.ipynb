DECISION TREE ALGORITHM - IRIS

```python
import pandas as pd
import sklearn.datasets as ds
import matplotlib.pyplot as pt
```

load the iris data sets

```python
data = ds.load_iris()
df = pd.DataFrame(data.data,columns = data.feature_names)
print(df.head())
y=data.target
print(y)
```

IMPORT DECISION TREE CLASSIFIER AND DEFINE ITS FUNCTION WITH MINIMUM SPLIT VALUE

```python
from sklearn.tree import DecisionTreeClassifier
md = DecisionTreeClassifier(min_samples_split=5)
md.fit(df,y)
```

VISUALIZATION - TREE

```python
from sklearn import tree
pt.figure(figsize = (20,20))
tree.plot_tree(md , filled = True)
pt.title("Decisionn Tree of Iris Data")
```
